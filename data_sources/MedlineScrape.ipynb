```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Wed Oct  2 02:10:06 2019

@author: gurdit.chahal
"""

#get dictonairy of drug names +urls to individual pages
import urllib.request
medline_url="https://medlineplus.gov/druginfo/herb_All.html"
#access url page for drugs
page=urllib.request.urlopen(medline_url)
import re
import pandas as pd

from bs4 import BeautifulSoup
soup=BeautifulSoup(page)
#print(soup.prettify())
```

```python
#extract everything between <a href= and >
web_pattern="(?<=<a href=)(.*?)>"
#grab everything between > and <
name_pattern="(?<=>)[^<]*"

#get all lines
lines=soup.find_all('li')
#initialize variables
drug_dict={}
url='unknown'
drug='unknown'
store_flag=False

#within lines look for website references
for line in lines:
    ref=str(line.find('a'))
    html=re.search(web_pattern,ref)
    if html:
        url=html.group(1).replace('"',"")
        #append https body to websites starting only with natural
        if url.startswith('natural'):
            url='https://medlineplus.gov/druginfo/'+url
            
    drug_name=re.search(name_pattern,ref)
  
    #only keep valid websites
    if url.startswith('https'):
        store_flag=True
    else:
        store_flag=False
    if bool(drug_name)&(store_flag):
        drug=drug_name.group()
    url=url.replace(' target=_blank','')
    drug_dict[drug]=url
#print(drug_dict)
```

```python
## NIH related herbs scrape
## Outputting Herb and 5 attributes in separate columns
drug_info2=[]
# Loop through drug dict looking for URL's with nccih
# Build list of drugs and page contents from the associated URL
for d in drug_dict.items():
    if d[1][8:13] == 'nccih':
        drug_url2=d[1]
        drug_page2=urllib.request.urlopen(drug_url2)
        drug_soup2=BeautifulSoup(drug_page2)
        #common names
        if drug_soup2.find(class_="field field-name-field-herb-names field-type-text field-label-inline clearfix") is None:
            common_name = ''
        else:
            common_name=drug_soup2.find(class_="field field-name-field-herb-names field-type-text field-label-inline clearfix").text
        #Background
        if drug_soup2.find(id="hed1") is None:
            background=''
        else:
            background=drug_soup2.find(id="hed1").next_sibling.next_sibling.text
        
        #We know
        if drug_soup2.find(id="hed2") is None:
            know=''
        else:
            know=drug_soup2.find(id="hed2").next_sibling.next_sibling.text
        #We learned
        if drug_soup2.find(id="hed3") is None:
            learn = ''
        else:
            learn=drug_soup2.find(id="hed3").next_sibling.next_sibling.text
        
        #Safety
        if drug_soup2.find(id="hed4") is None:
            safety=''
        else:
            safety=drug_soup2.find(id="hed4").next_sibling.next_sibling.text
        #refs
        if drug_soup2.find(class_="refs") is None:
            refs=''
        else:
            refs = drug_soup2.find(class_="refs").text 
                
        drug_info2.append([d[0],common_name,background,know,learn,safety])
        #print(d[0])
#print(drug_info2)
```

```python
## NIH related herbs scrape
## Outputting Herb and one attribute of all contents of its page
drug_info=[]
# Loop through drug dict looking for URL's with nccih
# Build list of drugs and page contents from the associated URL
for d in drug_dict.items():
    if d[1][8:13] == 'nccih':
        drug_url2=d[1]
        drug_page2=urllib.request.urlopen(drug_url2)
        drug_soup2=BeautifulSoup(drug_page2)
        main_content2=str(drug_soup2.find_all('li'))
        drug_info.append([d[0],main_content2])
        #print(d[0])
```
