```python
import csv
import pandas as pd
import os
```

### Combining herb disease files into one

```python
for filename in os.listdir('c:/selenium/named_files/'):
    print('https://www.symmap.org/detail/'+filename[0:9])
    #https://www.symmap.org/detail/SMMS00192
```

```python
herb_disease=[]
for filename in os.listdir('c:/selenium/named_files/'):
    print(filename)
    with open('c:/selenium/named_files/'+filename,encoding='utf8') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        for row in csv_reader:
            herb_disease.append([filename[:9],row[0],row[1],row[2],row[3],row[4],row[5],row[6],row[7],'https://www.symmap.org/detail/'+filename[0:9]])
herb_disease
```

```python
df = pd.DataFrame(herb_disease, columns =[
    'Herb id',
    'Disease id',
    'Disease name',
    'OMIM id',
    'Orphanet id',
  'P_value',
  'FDR(BH)',
  'FDR(Bonferroni)',
  'Relationship',
    'url']) 
df
df.to_csv ('./SymScrapeDisease1.csv', index =None, header=True) 
```

### Combining Symtoms file into one

```python
herb_symptom=[]
for filename in os.listdir('c:/selenium/symptom_files/'):
    print(filename)
    with open('c:/selenium/symptom_files/'+filename,encoding='utf8') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        for row in csv_reader:
            herb_symptom.append([filename[:9],row[0],row[1],row[2],row[3],'https://www.symmap.org/detail/'+filename[0:9]])
herb_symptom
```

```python
df2 = pd.DataFrame(herb_symptom, columns =[
    'Herb id',
    'MM symptom id',
    'MM symptom name',
    'MeSH tree numbers',
    'UMLS id',
    'url']) 
df2
df2.to_csv ('./SymScrapeSymptoms1.csv', index =None, header=True) 
```
