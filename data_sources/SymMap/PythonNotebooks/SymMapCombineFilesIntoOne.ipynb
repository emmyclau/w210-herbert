```python
import csv
import pandas as pd
import os
```

```python
herb_disease=[]
for filename in os.listdir('c:/selenium/named_files/'):
    print(filename)
    with open('c:/selenium/named_files/'+filename,encoding='utf8') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        for row in csv_reader:
            herb_disease.append([filename[:9],row[0],row[1],row[2],row[3],row[4],row[5],row[6],row[7]])
herb_disease
```

```python
df = pd.DataFrame(herb_disease, columns =[
    'Herb id',
    'Disease id',
    'Disease name',
    'OMIM id',
    'Orphanet id',
  'P_value',
  'FDR(BH)',
  'FDR(Bonferroni)',
  'Relationship']) 
df
df.to_csv ('./SymScrapeDisease1.csv', index =None, header=True) 
```
